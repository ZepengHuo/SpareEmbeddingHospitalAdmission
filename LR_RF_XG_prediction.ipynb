```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import f1_score
```

```python
import pandas as pd
```

```python
#options: output_file_h100_nonsparse     output_file_5


df_dense = pd.read_csv('output_file_h100_nonsparse', header=None, sep=' ')
df_sparse = pd.read_csv('output_file_5', header=None, sep=' ')
df_dense = df_dense.drop(columns=[0, 101])
df_sparse = df_sparse.drop(columns=[0, 101])
#df = df.drop(columns=[101])
```

```python
#labels:  1: admit    0: discharge
df_labels = pd.read_csv('space_separated_labels.csv', header=None, sep=' ')
df_labels = df_labels.drop(columns=[0,2])
df_labels = pd.get_dummies(df_labels).drop(columns='1_Discharge')
```

```python
#choose 1.sparse  2.dense

df = df_sparse
#df = df_dense

df
```

```python
clf_lr = LogisticRegression(random_state=0, solver='sag', multi_class='auto').fit(
    df.loc[0:90000,:].values.tolist(), df_labels.loc[0:90000,:].values.tolist())
```

```python
clf_rf = RandomForestClassifier(n_estimators=100, max_depth=100, random_state=0).fit(
    df.loc[0:90000,:].values.tolist(), df_labels.loc[0:90000,:].values.tolist())
```

```python
clf_xg = XGBClassifier().fit(
    df.loc[0:90000,:].values, df_labels.loc[0:90000,:].values)
```

```python
#options: clf_lr, clf_rf, clf_xg
clf = clf_xg

pred_list = clf.predict(df.loc[90001:100000,:].values.tolist())
pred_proba_list = clf.predict_proba(df.loc[90001:100000,:].values.tolist())
```

```python
true_list = df_labels.loc[90001:100000,:].values.tolist()
```

```python
accuracy_score(true_list, pred_list)
```

```python
roc_auc_score(true_list, pred_proba_list[:,1])
```

```python
f1_score(true_list, pred_list)
```
